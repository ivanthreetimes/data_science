## Second

```python
tb = pd.read_csv("data/tb.csv")
tb
```

```python
tb.describe()
```

```python
tb_tidy = pd.melt(tb, id_vars=["iso2","year"], var_name="gender_and_age", value_name="cases")
tb_tidy
```

```python
tb_tidy["asd"] = np.nan
```

```python
tb_tidy = tb_tidy.dropna()
tb_tidy
```

```python
parts = tb_tidy["gender_and_age"].str.extract("(\D)(\d+)(\d{2})", expand = True)
parts.columns = ["gender", "age_lower", "age_upper"]
parts["age"] = parts["age_lower"] + "-" + parts["age_upper"]
```

```python
tb_tidy = pd.concat([tb_tidy, parts], axis = 1)
tb_tidy
```

```python
tb_tidy = tb_tidy.drop(["gender_and_age", "age_lower", "age_upper"], axis=1)
tb_tidy
```

```python
tb_tidy = tb_tidy.dropna()
tb_tidy
```

```python
tb_tidy = tb_tidy.sort_values(by = ["iso2", "year", "gender", "age", "cases"])
```

```python
tb_tidy = tb_tidy.reset_index() #наименова колоната с индексите index
#tb_tidy = tb_tidy.reset_index(drop=True) #маха оригиналните индекси, слага нови - последователни и кръщава колоната index, над тях
tb_tidy
```

```python
del tb_tidy["index"] #трие оригиналните индекси под колоната index и слага новите...
tb_tidy
```

## Third

```python
temp_data = pd.read_csv("data/weather.csv")
temp_data
```

```python
temp_data.describe()
```

```python
temp_data = pd.melt(temp_data, id_vars=["id", "year", "month", "element"], var_name="day")
temp_data
```

```python
temp_data["day"] = temp_data["day"].str.slice(1).astype(np.int64)
temp_data
```

```python
temp_data = temp_data.pivot_table(index=["id", "year", "month", "day"], columns=["element"], values=["value"])
temp_data #Multi indexing. We can fix it by using reset.index as show below
```

```python
temp_data = temp_data.reset_index()
temp_data
```

```python
temp_data.columns
```

```python
temp_data.columns = ["id", "year", "month", "day", "tmax", "tmin"] #The columns are fixed now as well.
temp_data.columns
```

```python
new = pd.DataFrame()
new["year"] = temp_data["year"]
new["month"] = temp_data["month"]
new["day"] = temp_data["day"]
temp_data["date"] = pd.to_datetime(new)
del temp_data["year"], temp_data["month"], temp_data["day"] 
```

```python
temp_data = temp_data[["id", "date", "tmax", "tmin"]]
temp_data
```

```python
temp_data.drop("id", axis=1)
```
